```python
class calc:
  def __init__(self):
    print("welcome to calc")
  def add(self, a, b):
    return a + b
  def minus (self, a, b):
    return a - b
  def multi (self, a, b):
    return a * b
  def divide (self, a, b):
    return a / b
  def modules(self, a, b):
    return a % b
calcc = calc()
print(calcc.add(10, 10))
print(calcc.minus(10, 10))
print(calcc.multi(10, 10))
print(calcc.divide(10, 10))
print(calcc.modules(10, 10))
```

```python
class Vehicle:

    def __init__(self, make, model, year):

        self.make = make

        self.model = model

        self.year = year

    def start_engine(self):

        print("Starting engine...")
 
#derived class/child class

class Car(Vehicle):

    def __init__(self, make, model, year, num_doors):

        super().__init__(make, model, year)

        self.num_doors = num_doors

    def drive(self):

        print("Driving the car...")
 
vehicle = Vehicle("Toyota", "Corolla", 2022)

vehicle.start_engine()  # prints "Starting engine..."
 
car = Car("Tesla", "Model S", 2021, 4)

car.start_engine()      # prints "Starting engine..."
 
#child can call parent functions but parent cannot call child functions

#vehicle.drive()       # This will produce an error, as the drive() method is not defined for the Vehicle class

car.drive()
 
```
